# 6. Outliers and Replacing them

```python
plt.figure(figsize=(10,30))
data.boxplot(rot = 270)
```

```python
data['Length'] = np.where(data['Length'] < 0.25, 0.5, data['Length']) 
data['Diameter'] = np.where(data['Diameter'] < 0.17, 0.4, data['Diameter']) 
data['Height'] = np.where(data['Height'] < 0.01, 0.1, data['Height']) 
data['Height'] = np.where(data['Height'] > 0., 0.1, data['Height']) 
data['Whole weight'] = np.where(data['Whole weight'] > 2.1, 0.8, data['Whole weight'])
data['Shucked weight'] = np.where(data['Shucked weight'] > 0.9, 0.3, data['Shucked weight']) 
data['Viscera weight'] = np.where(data['Viscera weight'] > 0.47, 0.1, data['Viscera weight']) 
data['Shell weight'] = np.where(data['Shell weight'] > 0.6, 0.2, data['Shell weight']) 
data['Age'] = np.where(data['Age'] > 17, 11, data['Age']) 
data['Age'] = np.where(data['Age'] < 6, 11, data['Age']) 
```

```python
plt.figure(figsize=(10,20))
data.boxplot(rot = 270)
```

# 7. Encoding on Categorical values

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["Sex"]=le.fit_transform(data["Sex"])
```

# 8. Splitting into dependant and independant variables

```python
x = data.iloc[:, 0:8]
x
```

```python
y = data['Age']
y
```

# 9. Scaling the independant variables

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
x_scaled = mms.fit_transform(x)
```

```python
x_scaled
```

# 10. Splitting into training and testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.3, random_state = 1) 
```

```python
x_train
```

# 11. Building the model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

# 12. Training the model

```python
lr.fit(x_train, y_train)
```

# 13. Testing the model

```python
pred_test=lr.predict(x_test)
pred_test
```

# 14. Evaluation Metrics

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(pred_test, y_test)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```
